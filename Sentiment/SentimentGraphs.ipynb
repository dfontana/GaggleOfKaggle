```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('sentiment_days.csv')
```

```python
modes = data[['date','mode']]
modes = modes.copy()
sums = data[['date','Sum','Sum_QDAP']]
sums = sums.copy()
means = data[['date','Mean','Mean_QDAP']]
means = means.copy()
```

```python
modes['date'] = pd.to_datetime(modes['date'])
modes.set_index(modes['date'], inplace=True)
sums['date'] = pd.to_datetime(sums['date'])
sums.set_index(sums['date'], inplace=True)
means['date'] = pd.to_datetime(means['date'])
means.set_index(means['date'], inplace=True)
```

```python
# Graph of Sentiment by Mode
Moods1 = modes.groupby('mode')
MoodsMode = Moods1['mode'].count()
x = np.arange(3)
plt.bar(x, MoodsMode, color=['red','gray','blue'])
plt.xticks(x, MoodsMode.index)
plt.title('Sentiment Distribution by Mode')
plt.show()
```

```python
# Graph of Sentiment by Sum - Distribution
Moods2 = sums.groupby('Sum')
MoodsSum = Moods2['Sum'].count()
x = np.arange(2)
plt.bar(x, MoodsSum, color=['red','blue'])
plt.xticks(x, MoodsSum.index)
plt.title('Sentiment Distribution by Sum')
plt.show()
```

```python
# Graph of Sentiment by Sum - Numerical
plt.plot(sums.index, sums['Sum_QDAP'])
plt.title("Summed Sentiment by Day")
plt.rcParams['figure.figsize'] = [30, 20]
plt.show()
```

```python
# Graph of Sentiment by Mean - Distribution
Moods3 = means.groupby('Mean')
MoodsMean = Moods3['Mean'].count()
x = np.arange(2)
plt.bar(x, MoodsMean, color=['red','blue'])
plt.xticks(x, MoodsMean.index)
plt.title('Sentiment Distribution by Mean')
plt.show()
```

```python
# Graph of Sentiment by Mean - Numerical
plt.plot(means.index, means['Mean_QDAP'])
plt.title("Mean Sentiment by Day")
plt.rcParams['figure.figsize'] = [30, 10]
plt.show()
```

```python
from csv import reader
from csv import writer
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandasql as ps
```

```python
price = pd.read_csv("asx-equity-price.csv")
price.head()
```

```python
price["rel_difference_close_open"] = price["open"]-price["open"]
price["rel_difference_close_close"] = price["open"]-price["open"]

price.head()
```

```python
index1 = 0
index2 = 0
counter = 0 #checks if it is the first day the ticker is open

for ticker in price.ticker.unique():
    indexes = price.loc[price["ticker"] == ticker].index
    for row in indexes:
        if counter == 0: #first day for the given ticker
            index1 = row
            index2 = row
            counter+=1
        else:
            counter+=1
            index2 = row
            price.at[row, "rel_difference_close_open"] = price.iloc[index2]["open"] - price.iloc[index1]["close"]
            price.at[row, "rel_difference_close_close"] = price.iloc[index2]["close"] - price.iloc[index1]["close"]
            index1 = row
    counter = 0 #reset for the next ticker
            
```

```python
pysql = lambda q: ps.sqldf(q, globals())
q1 = "SELECT price.ticker, price.open, price.close, companies.industry FROM price JOIN companies ON price.ticker = companies.ticker"
joined = pysql(q1)
```

```python
joined
```

```python
q2 = "SELECT ticker, AVG(open) AS avg_open, AVG(close) AS avg_closed, industry FROM joined GROUP BY ticker" 
averaged = pysql(q2)
```

```python
averaged
```

```python
q3 = "SELECT ticker, avg_open, avg_closed, industry, (avg_open-avg_closed)/avg_closed AS rel_change FROM averaged" 
relative_change = pysql(q3)
```
